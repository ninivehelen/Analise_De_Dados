```python
dados = pd.read_csv("/content/drive/MyDrive/esus-vepi.LeitoOcupacao.csv")
```

**Dados soma total de confirmados em clinicas**`

```python
dados[["estado", "ocupacaoConfirmadoCli"]].groupby(["estado"]).count().head(50).sort_values(by="ocupacaoConfirmadoCli",ascending=False)
```

**Dados dos Estados. Soma total de confirmados em UTI**

```python
dados[["estado", "ocupacaoConfirmadoUti"]].groupby(["estado"]).count().head(50).sort_values(by="ocupacaoConfirmadoUti",ascending=False)
```

**Dados dos Municípios. Soma total de confirmados em Clinicas**

```python
dados[["municipio", "ocupacaoConfirmadoCli"]].groupby(["municipio"]).count().head(50).sort_values(by="ocupacaoConfirmadoCli",ascending=False)
```

**Dados dos Municípios. Soma total de confirmados em UTI**

```python
dados[["municipio", "ocupacaoConfirmadoUti"]].groupby(["municipio"]).count().head(50).sort_values(by="ocupacaoConfirmadoUti",ascending=False)
```

**Dados dos Estado. Soma total de alta em Clinicas**

```python
dados[["estado", "saidaConfirmadaAltas"]].groupby(["estado"]).count().sort_values(by='saidaConfirmadaAltas',ascending=False)
```

**Dados dos Municípios. Soma total de alta em Uti**

```python
dados[["municipio", "saidaConfirmadaAltas"]].groupby(["municipio"]).count().head(50).sort_values(by='saidaConfirmadaAltas',ascending=False)
```

**Dados dos Estados. Soma total de Obitos**

```python
dados[["estado", "saidaConfirmadaObitos"]].groupby(["estado"]).count().head(50).sort_values(by="saidaConfirmadaObitos",ascending=False)
```

**Dados dos Municípios. Soma total de Obitos**

```python
dados[["municipio", "saidaConfirmadaObitos"]].groupby(["municipio"]).count().head(50).sort_values(by="saidaConfirmadaObitos",ascending=False)
```
