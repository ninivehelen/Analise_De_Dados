````python
#Importando biblioteca pandas para leitura dos dados
import pandas as pd
````

````python
#Excluindo colunas que não serão ultilizadas na análise
df.drop(columns=['id_municipio',	'bairro_revenda',	'cep_revenda','endereco_revenda','cnpj_revenda','nome_estabelecimento','data_coleta','unidade_medida'],inplace=True)
df
````

````python
#Verificando se possui dados nulos
df.isnull().sum()
````

**Aplicando Limpeza e transformação dos dados**

````python
#importando biblioteca matematica 
import math
````

````python
#Calculando a media do preço venda para preencher nos dados ausentes
media = df["preco_compra"]. mean()
#Arredondando valor da media 
media2 = round(media, 4)
media2
````

````python
#Substituindo valor nulo no preço compra pela media calculada
df.update(df["preco_compra"].fillna(media2))
````

````python
#Vericando valores nulos novamente
df.isnull().sum()
````

````python
#Removendo valores nulos das colunas bandeira_revenda,preco_venda 
#Como esses valores so possuí uma unica linha com valores nulos, nao afetará a analise e a integridade dos dados
df.dropna(inplace=True)

````

````python
#Verificando novamente dados ausentes 
df.isnull().sum()
````

````python
#Selecionado apenas as linhas que serão ultilizadas na analise. 
values=[2019, 2020, 2021]
df_analise = df[df.ano.isin(values)]
````
