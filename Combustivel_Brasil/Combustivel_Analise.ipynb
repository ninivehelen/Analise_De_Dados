````python
#Importando biblioteca pandas para leitura dos dados
import pandas as pd
````

````python
#Ultilizando dados do csv criado nas etapas anteriores dados de 2019 a 2021
df_analise = pd.read_csv("/content/Combustivel3.csv")
df_analise
````

**Exploração dos dados**

````python
#Vericando dados da coluna UF
df_analise["sigla_uf"].unique()
````

````python
#Importando bibliotecas de criação de graficos
import matplotlib.pyplot as plt
import seaborn as sns
````

**Analisando dados sobre combustivel**

````python
#Grafico de preço do produto
produto = df_analise["produto"]
valor = df_analise["preco_venda"]
fig = plt.figure(figsize=(10,5))
sns.barplot(x=produto, y = valor, data=df_analise, palette="Paired")
plt.title('Preço do produto', fontsize=19);
````

***De acordo com o gráfico o Gás liquefeito de petróleo</br> é o gás com custo de preço mais caro no Brasil***

O gás liquefeito de petróleo, também chamado </br> de gás de petróleo liquefeito, </br>é uma mistura de gases de hidrocarbonetos </br> utilizado como combustível em aplicações </br>de aquecimento e veículos. O GLP ou GPL é a mistura </br>de gases condensáveis presentes no gás </br>natural ou dissolvidos no petróleo

````python
# Vericando dados da coluna ano
df_analise["ano"].unique()
````

````python
#Qual uf tem o preço de combustivel mais alto dos anos de 2019 a 2020
produto = df_analise["sigla_uf"]
valor = df_analise["preco_venda"]
fig = plt.figure(figsize=(10,5))
sns.barplot(x=produto, y = valor, data=df_analise, palette="Paired")
plt.title('Preço de combutivel por UF', fontsize=19);
````

***Roraima Tem um crescimento alto preço da venda de combustivel***

````python
#Crescimento geral da media de preço de compra combustivel no Brasil por ano 2019 a 2020
plt.figure(figsize=(15,5))
sns.lineplot(x=df_analise['ano'],y=df_analise['preco_compra'], color="blue")
plt.xlabel('Preços da venda de combustivel por ano')
plt.ylabel('anos')
plt.show()
````

***O preço da compra de combustivel no brasil ***

````python
#Crescimento geral da media de preço de venda combustivel no Brasil por ano 2019 a 2020
plt.figure(figsize=(15,5))
sns.lineplot(x=df_analise['ano'],y=df_analise['preco_venda'], color="green")
plt.xlabel('Preços da venda de combustivel por ano')
plt.ylabel('anos')
plt.show()
````

**Crescimento do preço da venda de combustivel no Brasil aumentou na metade de 2020**

**Analisando dados de preço da gasolina no Brasil**

````python
#Selecionando apenas dados sobre o preço da gasolina
df_gasolina = df_analise[df_analise["produto"] == "gasolina"]
````

````python
#Qual uf tem o preço de Gasolina mais alto dos anos de 2019 a 2020
produto = df_gasolina["sigla_uf"]
valor = df_gasolina["preco_venda"]
fig = plt.figure(figsize=(10,5))
sns.barplot(x=produto, y = valor, data=df_gasolina, palette="Paired")
plt.title('Preço da gasolina por UF', fontsize=19);
````

***O preço da venda de gasolina entre os estados estão bem proximos nos valores***

````python
#Crescimento geral da media de preço de compra de gasolinano Brasil por ano 2019 a 2020
plt.figure(figsize=(15,5))
sns.lineplot(x=df_gasolina['ano'],y=df_gasolina['preco_compra'], color="blue")
plt.xlabel('Preços da venda de combustivel por ano')
plt.ylabel('anos')
plt.show()
````

***O preço da compra de gasolina no Brasil teve um aumento a parti do ano de 2019***

````python
#Crescimento geral de  preço de venda de gasolina no Brasil por ano 2019 a 2020
plt.figure(figsize=(15,5))
sns.lineplot(x=df_gasolina['ano'],y=df_gasolina['preco_venda'], color="green")
plt.xlabel('Preços da venda de combustivel por ano')
plt.ylabel('anos')
plt.show()
````

***O preço da venda de gasolina no Brasil teve um aumento na metade do ano de 2020***
